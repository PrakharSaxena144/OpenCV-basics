### Image Processing

```python
# Image Transformation (shifting)
pic= cv2.imread("C:/Users/LENOVO/Desktop/imbatman.png")
cols= pic.shape[1]
rows= pic.shape[0]

M= np.float32([[1,0,150], [0,1,70]]) # Put here opp sign to get opp Transformation

shifted= cv2.warpAffine(pic, M, (cols, rows))
cv2.imshow('shifted', shifted)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# Rotation
pic= cv2.imread("C:/Users/LENOVO/Desktop/imbatman.png")
cols= pic.shape[1]
rows= pic.shape[0]
center= (cols/2, rows/2)
angle= 180
M= cv2.getRotationMatrix2D(center, angle, 1)

rotate= cv2.warpAffine(pic, M, (cols, rows))
cv2.imshow('rotated', rotate)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# Thresholding

# Read image (convert to grayscale for thresholding)
pic = cv2.imread("C:/Users/LENOVO/Desktop/imbatman.png", 0)

threshold_value = 120 # Pixels having intensity less than this are blacked out

# threshold() returns ret value and binary image
ret, binary_threshold = cv2.threshold(pic, threshold_value, 255, cv2.THRESH_BINARY)

cv2.imshow("binary", binary_threshold)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

### Image Filtering

```python
# Gaussian Blur
import cv2
pic= cv2.imread("C:/Users/LENOVO/Desktop/imbatman.png")

matrix= (7,7) # Level of blur; (1,1) --> Clear; (7,7) --> blured
blur= cv2.GaussianBlur(pic, matrix, 0)

cv2.imshow('blurred', blur)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# Median Blur
pic= cv2.imread("C:/Users/LENOVO/Desktop/imbatman.png")

kernal= 3
blur= cv2.medianBlur(pic, kernal)

cv2.imshow('blurred', blur)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# Bilateral Filtering
pic= cv2.imread("C:/Users/LENOVO/Desktop/imbatman.png")

dimpixel= 7
color= 100
space= 100
filter= cv2.bilateralFilter(pic, dimpixel, color, space)

cv2.imshow('filter', filter)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

### Feature Extraction
##### Canny Edge Detector

```python
# All pixel below Threshd1 are put 0 and larger than Threhd2 are put 1
# the func is abt to find edges by recognising black and white spots and extracting them as features
pic= cv2.imread("C:/Users/LENOVO/Desktop/imbatman.png")

thresholdval1= 50
thresholdval2= 100
canny= cv2.Canny(pic, thresholdval1, thresholdval2)

cv2.imshow('canny', canny)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

## Video

```python
# Play a video
import cv2

cap = cv2.VideoCapture("C:/Users/LENOVO/Videos/AI_for_good_video.mp4")

while cap.isOpened():
    ret, frame = cap.read()

    if not ret:
        print("Finished / Failed to read video")
        break

    cv2.imshow('Video', frame)

    # Use 25 for smoother playback (matches ~40fps). Use 1 for webcam.
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```

```python
# save a video in diff formate
import cv2

cap = cv2.VideoCapture("C:/Users/LENOVO/Videos/AI_for_good_video.mp4")

fourcc = cv2.VideoWriter_fourcc(*'XVID')
fps = 30
framesize = (720, 480)

out = cv2.VideoWriter('sample.avi', fourcc, fps, framesize)

while cap.isOpened():
    ret, frame = cap.read()
    
    if not ret:
        print("Finished / Failed to read video")
        break
    
    frame = cv2.resize(frame, framesize)   # ensure correct size
    out.write(frame)                       # save frame to file
    
    cv2.imshow('frame', frame)

    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
out.release()
cv2.destroyAllWindows()
```

# Applications

## Face Capture

```python
import cv2
import numpy as np

# https://github.com/opencv/opencv/tree/master/data/haarcascades
face_cascade = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')

videocapture = cv2.VideoCapture(0)
scale_factor = 1.3

while True:
    ret, pic = videocapture.read()
    
    if not ret:
        print("Failed to capture video")
        break

    faces = face_cascade.detectMultiScale(pic, scale_factor, 5)

    for (x, y, w, h) in faces:
        cv2.rectangle(pic, (x, y), (x + w, y + h), (255, 0, 0), 2)
        font = cv2.FONT_HERSHEY_SIMPLEX
        cv2.putText(pic, 'Me', (x, y), font, 2, (255, 255, 255), 2, cv2.LINE_AA)

    print("Number of faces found: {}".format(len(faces)))

    cv2.imshow('face', pic)

    k = cv2.waitKey(30) & 0xff
    if k == 27:      # ESC key to exit
        break

videocapture.release()
cv2.destroyAllWindows()
```
