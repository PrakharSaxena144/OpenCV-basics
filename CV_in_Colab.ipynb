```python
import cv2
from google.colab.patches import cv2_imshow
import numpy as np

# Load an image (example using a URL or a file you've uploaded)
# For a local file, ensure it's uploaded to your Colab environment first
!curl -o /content/Screenshot 2026-01-04 152159.png https://colab.research.google.com/img/colab_favicon_256px.png
img = cv2.imread('/content/Screenshot 2026-01-04 152159.png', cv2.IMREAD_UNCHANGED)

# Display the image using cv2_imshow
if img is not None:
    cv2_imshow(img)
else:
    print("Error: Image not loaded")

# Note: You don't need cv2.waitKey() or cv2.destroyAllWindows() with cv2_imshow in Colab.
```
